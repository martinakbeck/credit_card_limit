## 1.7 Descriptive Statiscal

```python
num_attributes = df1.select_dtypes(include=['int64', 'float64'])
cat_attributes = df1.select_dtypes(exclude=['int64', 'float64', 'datetime64[ns]'])
```

### 1.7.1 Numerical Attributes

```python
num_attributes.describe()
```

```python
# Central Tendency - mean, median
ct1 = pd.DataFrame (num_attributes.apply(np.mean)).T
ct2 = pd.DataFrame (num_attributes.apply(np.median)).T

# Dispersion - std, min, max, range, skew, kurtosis
d1 = pd.DataFrame(num_attributes.apply(np.std)).T
d2 = pd.DataFrame(num_attributes.apply(min)).T
d3 = pd.DataFrame(num_attributes.apply(max)).T
d4 = pd.DataFrame(num_attributes.apply(lambda x: x.max() - x.min())).T
d5 = pd.DataFrame(num_attributes.apply(lambda x: x.skew())).T
d6 = pd.DataFrame(num_attributes.apply(lambda x: x.kurtosis())).T

#concatenate
m = pd.concat ([d2, d3, d4, ct1, ct2, d1, d5, d6]).T.reset_index()
m.columns = ['attributes', 'min', 'max', 'range', 'mean', 'median', 'std', 'skew', 'kustosis']
m
```

## 2.2 Criação das Hipoteses

### 2.2.1 Hipóteses Cliente

**1.** Quanto maior a (renda anual) salário maior o valor investido

**2.** Quanto maior os gastos menor o valor investido

### 2.2.2 Hipóteses Banco

**1.** Quanto maior o número de cartões de crédito maior a dívida

**2.** Quanto maior o número de contas bancárias, maior o número de cartões de crédito

**3.** Quanto maior o número de cartões de crédito, maior é a quantidade de consultas ao crédito

### 2.2.3 Hipóteses Dívida

**1.** Quanto maior o número de pagamentos em atraso, maior o número de empréstimos

**2.** Quanto maior os dias de atrasos, maior a taxa a de utilização de crédito

**3.** Quanto maior a dívida atual, maior o número de emprétismos

## 2.3 Lista final de Hipóteses

**1.** Quanto maior a (renda anual) salário maior o valor investido

**2.** Quanto maior os gastos menor o valor investido

**3.** Quanto maior o número de cartões de crédito maior a dívida

**4.** Quanto maior o número de contas bancárias, maior o número de cartões de crédito

**5.** Quanto maior o número de cartões de crédito, maior é a quantidade de consultas ao crédito

**6.** Quanto maior o número de pagamentos em atraso, maior o número de empréstimos

**7.** Quanto maior os dias de atrasos, maior a taxa a de utilização de crédito

**8.** Quanto maior a dívida atual, maior o número de emprétismos

## 2.4 Feature Engineering

```python
mediana_idade = df2['idade'].median()
df2['idade'] = df2['idade'].map(lambda x: mediana_idade if x > 100 else x)

dict = {'Não' : 0, 'Sim' : 1}
df2[['investe_exterior', 'pessoa_polit_exp']] = df2[['investe_exterior', 'pessoa_polit_exp']].apply(lambda x: x.map(dict))

dict = {'Negar' : 0, 'Conceder' : 1}
df2['limite_adicional'] = df2['limite_adicional'].map(dict)
```
